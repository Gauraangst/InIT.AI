#### 15. Create a 2d array with 1 on the border and 0 inside (★☆☆)

````python
box = np.empty((10,10))
for i in range(10):
    for j in range(10):
        if i == 0  or i == 9 or j == 0 or j == 9:
            box[i][j] = 1
        else:
            box[i][j] = 0
print(box)

# O R -------------
arr = np.ones((5, 5), dtype=int)   # start with all 1s
arr[1:-1, 1:-1] = 0                # set inner elements to 0

print(arr)
````

#### 23. Create a custom dtype that describes a color as four unsigned bytes (RGBA) (★☆☆)

````python
rgba = np.dtype([
    ('r', 'u1'),
    ('g', 'u1'),
    ('b', 'u1'),
    ('a', 'u1')
])
pixels = np.array([
    (255, 0, 0, 255),   # Red
    (0, 255, 0, 128),   # Green (half transparent)
    (0, 0, 255, 64)     # Blue (faint)
], dtype=rgba)

print(pixels['r'])
print(pixels['g'])
print(pixels['b'])
print(pixels['a'])

````

#### 24. Multiply a 5x3 matrix by a 3x2 matrix (real matrix product) (★☆☆)

````python
matrix1 = np.random.randint(1,10, (5,3))
matrix2 = np.random.randint(1,10, (3,2))


print(matrix1)
print(matrix2)
print("-----------------------------------")
# there are multiple ways to multiply two matrices
print(matrix1 @ matrix2)
print(matrix1.dot(matrix2))
print(np.dot(matrix1, matrix2))
print(np.matmul(matrix1, matrix2))
````
